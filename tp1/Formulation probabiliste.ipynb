```python
%pylab inline
import matplotlib.pyplot as plot
import numpy as np
x = np.random.randn(100000)   # Génère 100000 échantillons d'une Gaussienne N(x|0,1)
dummy = plot.hist(x,100)
print ("moyenne=",mean(x))
print ("variance=",var(x))
```

```python
y = 2*x - 1 # Transforme les échantillons linéairement
dummy = plot.hist(y,100)
print ("moyenne=",mean(y))    # Devrait être autour de 2*E[x] -1 = -1
print ("variance=",var(y))    # Devrait êter autour de 2**2 * var[x] = 4
```

```python
mu = np.array([1,2])
sigma = np.array([[2,0],[0,1]])
y = np.random.multivariate_normal(mu,sigma,100000).T # Génère 100000 échantillons d'une Gaussienne N(x|[1,2]^T,[[2,0],[0,1]]) en 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print("moyenne=",y.mean(axis=1))
print("covariance=\n",cov(y))
```

```python
mu = np.array([1,2])
sigma = np.array([[1.5,1],[1,1.5]])
y = np.random.multivariate_normal(mu,sigma,100000).T # Génère 100000 échantillons d'une Gaussienne N(x|[1,2]^T,[[1.5,1],[1,1.5]]) en 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print("moyenne=",y.mean(axis=1))
print("covariance=\n",np.cov(y))
```

```python
mu = np.array([2,-1])
A = np.array([[3,0],[0,0.5]])
y = np.dot(A,x) + mu.reshape((2,1))
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = cm.Greys)
axis('equal')
print("moyenne=",y.mean(axis=1))
print('devrait être',mu)    # E[A x + mu] = A E[x]+mu = A(0)+mu = mu
print("covariance=\n",np.cov(y))
print('devrait être\n',np.dot(A,A.T))   # cov[A x + mu] = A cov[x] A^T = A I A^T = A A^T
```
